# Modeling

### Feature, Label

```python
df.head()
```

```python
df.columns
```

```python
X = df.loc[:, df.columns != 'Class']
y = df.loc[:, df.columns == 'Class']
print(X.shape, y.shape)
```

### Train-set/ Test-set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3000)
```

```python
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Model Training

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 500, max_depth = 10)
model.fit(X_train, y_train)
```

```python
# train data score
model.score(X_train, y_train)
```

```python
# test data
model.predict(X_test)
```

```python
# test data score
model.score(X_test, y_test)
```

```python
# test data probability
model.predict_proba(X_test)
```

### + Feature importance

```python
model.feature_importances_
```

```python
importances = model.feature_importances_
indices_sorted = np.argsort(importances)

fidf = pd.DataFrame()
fidf['variable'] = X.columns[indices_sorted]
fidf['importances'] =  importances[indices_sorted]
```

```python
import plotly.express as px
fig = px.bar(fidf, x = 'variable', y ='importances')
fig.show()
```

```python
top_n = 5
fig = px.bar(fidf.iloc[-top_n:], x = 'variable', y ='importances')
fig.show()
```
